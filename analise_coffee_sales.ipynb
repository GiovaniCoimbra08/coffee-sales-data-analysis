```python
df["date"] = pd.to_datetime(df["date"])
df["datetime"] = pd.to_datetime(df["datetime"])
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["datetime"] = pd.to_datetime(df["datetime"], format="mixed", errors="coerce")
```

```python
df.info()
```

```python
df[["date", "datetime"]].isnull().sum()
```

```python
df["ano"] = df["date"].dt.year
df["mes"] = df["date"].dt.month
df["dia"] = df["date"].dt.day
df["hora"] = df["datetime"].dt.hour
df["dia_semana"] = df["date"].dt.day_name()
```

```python
df["coffee_name_clean"] = df["coffee_name"].str.strip().str.lower()
df["cash_type"] = df["cash_type"].str.strip().str.lower()
```

```python
df.head()
```

```python
df["coffee_name_clean"].value_counts()
```

```python
df["coffee_name_formatado"] = df["coffee_name_clean"].str.title()
```

```python
df[["coffee_name", "coffee_name_clean", "coffee_name_formatado"]].head(10)
```

```python
resumo_cafe = df.groupby("coffee_name_formatado").agg(
    qtd_vendas=("coffee_name_formatado", "count"),
    faturamento_total=("money", "sum"),
    ticket_medio=("money", "mean")
).sort_values(by="faturamento_total", ascending=False)

resumo_cafe
```

```python
resumo_cafe = resumo_cafe.reset_index()

resumo_cafe["faturamento_total"] = resumo_cafe["faturamento_total"].round(2)
resumo_cafe["ticket_medio"] = resumo_cafe["ticket_medio"].round(2)

resumo_cafe.head(10)
```

```python
resumo_hora = df.groupby("hora").agg(
    qtd_vendas=("hora", "count"),
    faturamento_total=("money", "sum"),
    ticket_medio=("money", "mean")
).reset_index().sort_values(by="hora")

resumo_hora["faturamento_total"] = resumo_hora["faturamento_total"].round(2)
resumo_hora["ticket_medio"] = resumo_hora["ticket_medio"].round(2)

resumo_hora
```

```python
resumo_hora.sort_values(by="faturamento_total", ascending=False).head(10)
```

```python
resumo_dia_semana = df.groupby("dia_semana").agg(
    qtd_vendas=("dia_semana", "count"),
    faturamento_total=("money", "sum"),
    ticket_medio=("money", "mean")
).reset_index()

resumo_dia_semana["faturamento_total"] = resumo_dia_semana["faturamento_total"].round(2)
resumo_dia_semana["ticket_medio"] = resumo_dia_semana["ticket_medio"].round(2)

resumo_dia_semana.sort_values(by="faturamento_total", ascending=False)
```

```python
df["periodo_dia"] = pd.cut(
    df["hora"],
    bins=[5, 12, 15, 18, 24],
    labels=["Manhã", "Almoço/Início da tarde", "Tarde", "Noite"],
    right=False
)
```

```python
resumo_periodo = df.groupby("periodo_dia").agg(
    qtd_vendas=("periodo_dia", "count"),
    faturamento_total=("money", "sum"),
    ticket_medio=("money", "mean")
).reset_index()

resumo_periodo["faturamento_total"] = resumo_periodo["faturamento_total"].round(2)
resumo_periodo["ticket_medio"] = resumo_periodo["ticket_medio"].round(2)

resumo_periodo.sort_values(by="faturamento_total", ascending=False)
```

```python
resumo_periodo = df.groupby("periodo_dia").agg(
    qtd_vendas=("periodo_dia", "count"),
    faturamento_total=("money", "sum"),
    ticket_medio=("money", "mean"),
    horas_ativas=("hora", "nunique")
).reset_index()

resumo_periodo["faturamento_total"] = resumo_periodo["faturamento_total"].round(2)
resumo_periodo["ticket_medio"] = resumo_periodo["ticket_medio"].round(2)

resumo_periodo["vendas_media_por_hora"] = (
    resumo_periodo["qtd_vendas"] / resumo_periodo["horas_ativas"]
).round(2)

resumo_periodo["faturamento_medio_por_hora"] = (
    resumo_periodo["faturamento_total"] / resumo_periodo["horas_ativas"]
).round(2)

resumo_periodo.sort_values(by="faturamento_medio_por_hora", ascending=False)
```

```python
df["card_status"] = df["card"].copy()

df.loc[(df["cash_type"] == "cash") & (df["card_status"].isna()), "card_status"] = "Não se aplica"

df.loc[(df["cash_type"] == "card") & (df["card_status"].isna()), "card_status"] = "Não informado"
```
